<font color = "maroon" size=3px><b>1.4 lets derive new variable IsPool from PoolQc, which states wether house contains Pool or not </b></font>

```python
# Function For Creating new Feature

def create(val):
    if val =="Ex" or val == "Fa" or val == "Gd":
        return(1)
    else:
        return(0)
```

```python
# Creating isPool Feature in train and test

for df in [train,test]:
    df['isPool'] = df.PoolQC.apply(create)

train.isPool.unique()
```

<font color = "maroon" size=4px><b>lets remove Highly Correlated Features </b></font>

```python
# Creating a function through which we can select highly correlated features and dropping the first feature that is correlated with any other feature

def correlation(dataset, threshold):
    col_corr = set() # Set of all names of correlated columns
    corr_matrix = dataset.corr() # Correlation Matrix
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:    # we are interested in absolute coeff value
                    
                colname = corr_matrix.columns[i] # getting the name of columns
                col_corr.add(colname)
    return(col_corr)
```

```python
from xgboost import XGBRegressor

def objective(trial, X, y, random_state , n_splits, n_jobs, early_stopping_rounds):

# Param Space
    params = {
            "verbosity": 0,  # 0 (silent) - 3 (debug)
            'objective': 'reg:squarederror',
            'tree_method':'gpu_hist',# Use GPU acceleration
            'predictor': 'gpu_predictor',
            'eval_metric' : 'rmse',
            "seed": random_state,
            'n_jobs': n_jobs,
            "alpha": trial.suggest_loguniform("alpha", 0.01, 10.0),
            "lambda": trial.suggest_loguniform("lambda", 1e-8, 10.0),
            "gamma": trial.suggest_loguniform("lambda", 1e-8, 10.0),
            "colsample_bytree": trial.suggest_loguniform("colsample_bytree", 0.2, 0.6),
            "subsample": trial.suggest_loguniform("subsample", 0.5, 0.8),
            "learning_rate": trial.suggest_loguniform("learning_rate", 0.005, 0.05),
            'n_estimators': 4000,
            'max_depth': trial.suggest_int("max_depth", 1,20,log = True),
            "min_child_weight": trial.suggest_loguniform("min_child_weight", 1, 300),
        }


    # Call Back For pruning unpromising trails
    pruning_callback = optuna.integration.XGBoostPruningCallback(trial, "validation_0-rmse")
    model = XGBRegressor(**params)
    
    # Defining Kfold Loop Inputs
    X_values = X.values
    y_values = y.values
    metric = []
    
    # Cross-Validation Loop:

    for train_index, test_index in kf.split(X_values):
        X_train, X_test = X_values[train_index, :], X_values[test_index, :]
        y_train, y_test = y_values[train_index], y_values[test_index]
        model.fit(
            X_train,
            y_train,
            eval_set=[(X_test, y_test)],
            eval_metric="rmse",
            verbose=0,
            callbacks=[pruning_callback],
            early_stopping_rounds=early_stopping_rounds,
        )
        y_pred = model.predict(X_test)
        
        # Storing Each Prediction accuracy of Kfold in List
        metric.append(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
        
    metric = np.mean(metric)
    return (metric)
```

```python
%%time

# Creating Study

study_xgb = optuna.create_study(direction="minimize",pruner=optuna.pruners.HyperbandPruner())
study_xgb.optimize(
    lambda trial: objective(
        trial,
        features_GLLM,
        target,
        random_state=101,
        n_splits=10,
        n_jobs=-1,
        early_stopping_rounds=40,
    ),
    n_jobs=-1,
    n_trials=100
)
```

```python
# Calculating the pruned and completed trials

pruned_trials = [t for t in study_xgb.trials if t.state == optuna.trial.TrialState.PRUNED]
complete_trials = [t for t in study_xgb.trials if t.state == optuna.trial.TrialState.COMPLETE]

print("  Number of finished trials: ", len(study_xgb.trials))
print("  Number of pruned trials: ", len(pruned_trials))
print("  Number of complete trials: ", len(complete_trials))
```
